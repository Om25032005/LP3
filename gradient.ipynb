```python
import matplotlib.pyplot as plt

# Function
def f(x):
    return (x + 3)**2   # given function

# Derivative of the function
def df(x):
    return 2 * (x + 3)  # derivative of (x + 3)^2 is 2(x + 3)

# Gradient Descent Implementation
x = 2                 # starting point
learning_rate = 0.1   # step size
max_iter = 100        # number of iterations

x_history = []        # to store x values for plotting

print("Starting Gradient Descent...")
print("Initial x =", x)

for i in range(max_iter):
    grad = df(x)                 # compute gradient
    x = x - learning_rate * grad # update rule
    x_history.append(x)

    print(f"Iteration {i+1}: x = {x:.6f}, f(x) = {f(x):.6f}")

print("\nFinal Result:")
print("Local minimum occurs at x =", x)
print("Minimum value =", f(x))

# Plotting x values over iterations
plt.plot(x_history)
plt.xlabel("Iteration")
plt.ylabel("x value")
plt.title("Gradient Descent Convergence")
plt.grid(True)
plt.show()
```
